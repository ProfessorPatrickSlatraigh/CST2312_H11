## Opening Bridge: From Tuples to Dictionaries

In the previous session we worked with **tuples** — ordered, immutable sequences accessed by integer index position.

```
t = ('TX', 'Texas', 'Austin')
print(t[0])   # 'TX'
print(t[2])   # 'Austin'
```

Index-based access works, but it requires us to *remember which position holds which field*.

A **dictionary** solves that problem by letting us look up values using meaningful **keys** instead of integer positions.


````python
# Tuple: access by integer position
state_tuple = ('TX', 'Texas', 'Austin')
print("Abbreviation:", state_tuple[0])
print("State name  :", state_tuple[1])
print("Capital     :", state_tuple[2])
````

````python
# Dictionary: access by key (preview -- we will build this properly below)
state_dict = {'abbr': 'TX', 'name': 'Texas', 'capital': 'Austin'}
print("Abbreviation:", state_dict['abbr'])
print("State name  :", state_dict['name'])
print("Capital     :", state_dict['capital'])
````

| Structure | Index type | Example access | Duplicate indices allowed? |
|-----------|-----------|----------------|---------------------------|
| `list` | Integer (0-based) | `my_list[2]` | Yes |
| `dict` | Any immutable type (key) | `my_dict['TX']` | Keys: **No**. Values: Yes. |

## Segment 1 — Dictionary Fundamentals

A dictionary is a **mapping** between a set of **keys** and a set of **values**.  
Each key maps to exactly one value. The key-value pair is called an **item**.

> *"A dictionary is like a list, but more general. In a list, the index positions  
> have to be integers; in a dictionary, the indices can be (almost) any type."*  
> — Severance, *Python for Everybody*

### 1.1 — Three Ways to Create a Dictionary

#### Pattern A: Empty dict, then bracket assignment


````python
# Create an empty dictionary with the built-in dict() function
eng2sp = dict()
print(eng2sp)          # {} -- curly brackets represent an empty dictionary
print(type(eng2sp))
````

````python
# Add key-value pairs one at a time using bracket notation
eng2sp['one']   = 'uno'
eng2sp['two']   = 'dos'
eng2sp['three'] = 'tres'
print(eng2sp)
````

#### Pattern B: Literal notation `{ key: value, ... }`


````python
# Construct the dictionary in a single literal expression
eng2sp = {'one': 'uno', 'two': 'dos', 'three': 'tres'}
print(eng2sp)
````

#### Pattern C: `dict()` constructor variants

```
dict(**kwarg)              # keyword arguments
dict(mapping, **kwarg)     # from an existing mapping
dict(iterable, **kwarg)    # from an iterable of (key, value) pairs
```


````python
# Keyword arguments -- keys become the argument names
d = dict(red=1, blue=2, green=3)
print(d)
````

````python
# From an iterable of (key, value) tuples
d = dict([('red', 1), ('blue', 2), ('green', 3)])
print(d)
````

````python
# Using zip() to pair two lists into key-value pairs
d = dict(zip(['red', 'green', 'blue'], [1, 2, 3]))
print(d)
````

### 2.4 — The Histogram Pattern (counting characters in a string)

**Approach 1: if/else with the `in` operator**  
Check whether the character is already in the dictionary;  
if not, initialise it to 1; otherwise, increment.


````python
# Approach 1 -- if/else
# word_str   : a string to analyse
# histo_dict : histogram dictionary  (character -> count)
# char_str   : one character from the string

word_str   = 'brontosaurus'
histo_dict = dict()

for char_str in word_str:
    if char_str not in histo_dict:
        histo_dict[char_str] = 1
    else:
        histo_dict[char_str] = histo_dict[char_str] + 1

print(histo_dict)
````

**Approach 2: `get()` with a default value**

`dict.get(key, default)` returns the value for *key* if it exists,  
or *default* if it does not.  This collapses the four-line if/else into one line.


````python
# Demonstrate get() with a known key and an absent key
counts_dict = {'chuck': 1, 'annie': 42, 'jan': 100}
print(counts_dict.get('jan', 0))    # key exists  -> returns 100
print(counts_dict.get('tim', 0))    # key absent  -> returns 0
# compare: counts_dict['tim'] would raise a KeyError
````

````python
# Approach 2 -- get() one-liner  (same result, fewer lines)
word_str   = 'brontosaurus'
histo_dict = dict()

for char_str in word_str:
    histo_dict[char_str] = histo_dict.get(char_str, 0) + 1

print(histo_dict)
````

> **The `get()` idiom is idiomatic Python and appears frequently in analytics code.  
> Learn to recognise and write it fluently.**

## Segment 3 — Text Wrangling and Word Counting

We now move from counting characters to counting **words** in a multi-line text string.  
Before counting, we must **normalise** the text to avoid treating `"Brontosaurus"`,  
`"brontosaurus"`, and `"brontosaurus."` as three distinct words.

---


### 3.1 — The Source Text

We use a short passage embedded directly in the notebook as a docstring.  
No file upload is required.


````python
text_str = """
Brontosaurus are not good burgers.
Perhaps I will just have the fries.
Are they as good as you say they are?
Brontosaurus burgers are not for the faint of heart.
The fries, however, are remarkably good.
"""

print(type(text_str))
print(repr(text_str[:60]))   # inspect leading characters including \n
````

---

### 3.2 — Basic Word Count (lowercase only)

Step 1: convert to lowercase.  
Step 2: split on whitespace to get a list of tokens.  
Step 3: count with `get()`.


````python
# Basic word count -- lowercase, split on whitespace
word_lst   = text_str.lower().split()     # .split() with no argument splits on any whitespace
histo_dict = dict()

for word_str in word_lst:
    histo_dict[word_str] = histo_dict.get(word_str, 0) + 1

print(histo_dict)
````

---

### 3.3 — Removing Punctuation

The `string` module provides a string of all punctuation characters.  
`str.translate()` with `str.maketrans()` removes them efficiently.


````python
import string
print('Punctuation characters:', string.punctuation)
````

````python
# Method A: loop with str.replace()
# Works correctly but iterates once per punctuation character
word_lst   = text_str.lower().replace('\n', ' ').split(' ')
histo_dict = dict()

for word_str in word_lst:
    for character in string.punctuation:
        word_str = word_str.replace(character, '')
    if word_str:                                     # skip empty strings created by split
        histo_dict[word_str] = histo_dict.get(word_str, 0) + 1

print(histo_dict)
````

````python
# Method B: str.translate() -- more concise and efficient
# line.translate(str.maketrans('', '', delete_str))
# deletes every character in delete_str from the string

histo_dict = dict()

for line in text_str.split('\n'):
    line  = line.strip()
    line  = line.translate(str.maketrans('', '', string.punctuation))
    line  = line.lower()
    words = line.split()
    for word in words:
        histo_dict[word] = histo_dict.get(word, 0) + 1

print(histo_dict)
````

````python
# Filter: print only words that appear more than once in the Brontosaurus text
for key in histo_dict:
    if histo_dict[key] > 1:
        print(key, histo_dict[key])
````

### 3.5 — Sorting a Dictionary

**Sort by key** using `sorted(d.items())`.  
**Sort by value** by swapping (value, key) and then sorting.


````python
# Sort by key (alphabetical order)
d = {'a': 10, 'b': 1, 'c': 22}
for k, v in sorted(d.items()):
    print(k, v)
````

````python
# Sort by value (ascending): construct (value, key) tuples then sort
tmp = []
for k, v in d.items():
    tmp.append((v, k))
tmp = sorted(tmp)
print(tmp)
````

````python
# Sort by value (descending): one-liner using list comprehension
print(sorted([(v, k) for k, v in d.items()], reverse=True))
````

````python
# Apply to the Brontosaurus word count: top 5 most frequent words
top5 = sorted([(v, k) for k, v in histo_dict.items()], reverse=True)[:5]
for count, word in top5:
    print(f'{word:<15} {count}')
````

## Segment 4 — Nested Data Structures

Dictionary values can themselves be **complex types** — tuples or even other dictionaries.  
This section builds a U.S. state lookup structure in two forms and compares them.

---


### 4.1 — Source Data: Three Parallel Lists

We begin with three lists that contain related data for each state in the same order.  
They are analogous to columns in a relational table.


````python
st_abbr_lst = [
    "AL","AK","AZ","AR","CA","CO","CT","DE","FL","GA",
    "HI","ID","IL","IN","IA","KS","KY","LA","ME","MD",
    "MA","MI","MN","MS","MO","MT","NE","NV","NH","NJ",
    "NM","NY","NC","ND","OH","OK","OR","PA","RI","SC",
    "SD","TN","TX","UT","VT","VA","WA","WV","WI","WY"
]

st_name_lst = [
    "Alabama","Alaska","Arizona","Arkansas","California","Colorado",
    "Connecticut","Delaware","Florida","Georgia","Hawaii","Idaho",
    "Illinois","Indiana","Iowa","Kansas","Kentucky","Louisiana",
    "Maine","Maryland","Massachusetts","Michigan","Minnesota",
    "Mississippi","Missouri","Montana","Nebraska","Nevada",
    "New Hampshire","New Jersey","New Mexico","New York",
    "North Carolina","North Dakota","Ohio","Oklahoma","Oregon",
    "Pennsylvania","Rhode Island","South Carolina","South Dakota",
    "Tennessee","Texas","Utah","Vermont","Virginia","Washington",
    "West Virginia","Wisconsin","Wyoming"
]

st_capital_lst = [
    "Montgomery","Juneau","Phoenix","Little Rock","Sacramento",
    "Denver","Hartford","Dover","Tallahassee","Atlanta","Honolulu",
    "Boise","Springfield","Indianapolis","Des Moines","Topeka",
    "Frankfort","Baton Rouge","Augusta","Annapolis","Boston",
    "Lansing","St. Paul","Jackson","Jefferson City","Helena",
    "Lincoln","Carson City","Concord","Trenton","Santa Fe",
    "Albany","Raleigh","Bismarck","Columbus","Oklahoma City",
    "Salem","Harrisburg","Providence","Columbia","Pierre",
    "Nashville","Austin","Salt Lake City","Montpelier","Richmond",
    "Olympia","Charleston","Madison","Cheyenne"
]

print(f"Lengths: abbr={len(st_abbr_lst)}, names={len(st_name_lst)}, capitals={len(st_capital_lst)}")
````

---

### 4.2 — Form A: Dictionary of Tuples

Use `zip()` to pair names with capitals, then `dict()` to map abbreviations to those pairs.


````python
# Step 1: zip state names and capitals into a list of tuples
state_info = list(zip(st_name_lst, st_capital_lst))

# Inspect the first three entries
for item in state_info[:3]:
    print(item)
````

````python
# Step 2: create the dictionary  (abbreviation -> (name, capital))
state_dict_tuples = dict(zip(st_abbr_lst, state_info))

# Inspect a few entries
for abbr in ['CA', 'TX', 'NY']:
    print(abbr, '->', state_dict_tuples[abbr])
````

````python
# Lookup functions using .get() with a descriptive default
def get_state_name(abbr):
    """Return the full state name for a two-letter abbreviation."""
    return state_dict_tuples.get(abbr, ('Abbreviation not found', ''))[0]

def get_state_capital(abbr):
    """Return the capital city for a two-letter abbreviation."""
    return state_dict_tuples.get(abbr, ('', 'Capital not found'))[1]

# Example usage
abbr = 'TX'
print(f'The state name for {abbr} is {get_state_name(abbr)}.')
print(f'The capital of {abbr} is {get_state_capital(abbr)}.')
````

---

### 4.3 — Form B: Dictionary of Dictionaries

Replace the tuple value with a subordinate dictionary that names each field explicitly.  
Use a **dict comprehension** to build the structure in one expression.


````python
# Dict comprehension: abbreviation -> {'Name': ..., 'Capital': ...}
state_dict = {
    abbr: {'Name': name, 'Capital': capital}
    for abbr, name, capital in zip(st_abbr_lst, st_name_lst, st_capital_lst)
}

# Inspect the full entry for one state
print(state_dict['TX'])
````

````python
# Nested key access: outer key then inner key
abbr = 'TX'
print(f'The state name for {abbr} is {state_dict[abbr]["Name"]}.')
print(f'The capital of {abbr} is {state_dict[abbr]["Capital"]}.')
````

---

### 4.4 — Comparing the Two Forms

| | Dictionary of Tuples | Dictionary of Dictionaries |
|---|---|---|
| **Value access** | `d['TX'][0]` (by position) | `d['TX']['Name']` (by named key) |
| **Self-documenting?** | No — reader must know index meaning | Yes — key names describe the field |
| **Adding a field later** | Requires restructuring the tuple | Add a new key to the inner dict |
| **Best when** | Structure is fixed and small | Structure may grow; readability matters |

> **Design principle:** Prefer a dictionary of dictionaries when field names improve clarity  
> or when the structure is likely to evolve.

## Segment 5 — Set Operations on Dictionaries

Python dictionary `.keys()` and `.items()` views support **set operations**  
because keys are unique and immutable (like set members).

> *Analogy for students who have seen SQL:*  
> These operations mirror JOIN types in relational databases.

| Operator | On `.keys()` | On `.items()` | SQL analogy |
|----------|-------------|---------------|-------------|
| `&` Intersection | Keys in **both** dicts | `(k, v)` tuples in both | INNER JOIN on key AND value |
| `|` Union | **All** unique keys | All unique tuples | FULL OUTER JOIN |
| `-` Difference | Keys in left but **not** right | Tuples in left, not right | LEFT ANTI JOIN |
| `^` Symmetric Difference | Keys in **exactly one** dict | Tuples in exactly one dict | Non-overlapping rows |

---


### 5.1 — Setup: Three Example Dictionaries


````python
d1 = {'a': 1, 'b': 2, 'c': 3}
d2 = {'b': 2, 'c': 4, 'd': 5}
d3 = {'b': 2, 'c': 19, 'd': 5, 'e': 7}

print('d1:', d1)
print('d2:', d2)
print('d3:', d3)
````

````python
# Keys and items views -- note their types
print(type(d1.keys()))    # dict_keys  -- supports set operations
print(type(d1.items()))   # dict_items -- supports set operations
print(list(d1.keys()))
print(list(d1.items()))
````

---

### 5.2 — Intersection (`&`)

Extracts what is **common to both** dictionaries.


````python
# Keys common to d1 and d2
intersection_keys = d1.keys() & d2.keys()
print('Intersection of keys :', intersection_keys)
print('Type                  :', type(intersection_keys))   # set
````

````python
# Items: BOTH key AND value must match
# Note: ('c', 3) is in d1 and ('c', 4) is in d2 -- keys match but values differ, so excluded
intersection_items = d1.items() & d2.items()
print('Intersection of items:', intersection_items)
# Only ('b', 2) appears in both d1 and d2 with the same value
````

````python
# Build a new dictionary from the intersection of items
intersection_dict = dict(d1.items() & d2.items())
print('Intersection dict    :', intersection_dict)
````

````python
# Try with d2 and d3 -- more than one common item
print(dict(d2.items() & d3.items()))
````

---

### 5.3 — Union (`|`)

Extracts **all unique keys** from both dictionaries.


````python
# Union of keys
union_keys = d1.keys() | d2.keys()
print('Union of keys :', union_keys)
````

````python
# Union of items -- tuples with conflicting values for the same key are BOTH included in the set
union_items = d1.items() | d2.items()
print('Union of items:', union_items)
# Both ('c', 3) and ('c', 4) appear -- the set cannot deduplicate by key alone
````

````python
# WARNING: converting an item union to a dict when there are key collisions
# loses one of the conflicting values -- which one is retained is unspecified
union_dict = dict(d1.items() | d2.items())
print('Union dict (collision for "c"):', union_dict)
# 'c' will map to either 3 or 4 -- do not rely on this behaviour
````

---

### 5.4 — Difference (`-`)

Subtracts the **right-side** elements from the **left-side** set.  
Order matters: `d1 - d2` is not the same as `d2 - d1`.


````python
# Keys in d1 that are NOT in d2
difference_keys = d1.keys() - d2.keys()
print('d1 - d2 keys:', difference_keys)     # {'a'}

# Keys in d2 that are NOT in d1
print('d2 - d1 keys:', d2.keys() - d1.keys())   # {'d'}
````

````python
# Items in d1 not in d2
difference_items = d1.items() - d2.items()
print('d1 - d2 items:', difference_items)

# Build a dict from the difference -- safe only if the result has unique keys
difference_dict = dict(d1.items() - d2.items())
print('Difference dict:', difference_dict)
````

---

### 5.5 — Symmetric Difference (`^`)

Items that appear in **exactly one** of the two dictionaries.  
Equivalent to the union minus the intersection.


````python
# Keys in exactly one of d1 or d2
sym_diff_keys = d1.keys() ^ d2.keys()
print('Symmetric difference of keys :', sym_diff_keys)
````

````python
# Items in exactly one of d1 or d2
sym_diff_items = d1.items() ^ d2.items()
print('Symmetric difference of items:', sym_diff_items)
````

````python
# Note: ('c', 3) from d1 and ('c', 4) from d2 are both included (neither is in the other)
# Converting to dict loses one -- use with caution
sym_diff_dict = dict(d1.items() ^ d2.items())
print('Symmetric difference dict    :', sym_diff_dict)
````

## Debugging Reference

As data sets grow, inspecting every value by hand becomes impractical.  
Use these strategies when working with dictionaries and text data.

**Scale down the input.**  
If the program reads a text file, start with the first 10 lines or the smallest example  
that still reproduces the error. Reduce `n` to find the failure point, then increase it.

**Check summaries and types.**  
Rather than printing the entire dictionary, print `len(d)`, `type(v)`,  
or a slice of the sorted items. A value of the wrong type is a common runtime error.

```python
print(type(histo_dict))       # confirm it is a dict
print(len(histo_dict))        # confirm it has the expected number of keys
print(list(histo_dict.items())[:5])   # inspect only the first five items
```

**Write self-checks (sanity checks).**  
After building a word-count dictionary, verify that the total count equals the number of tokens:

```python
assert sum(histo_dict.values()) == len(word_lst), "Count mismatch -- check the loop"
```

**Write consistency checks.**  
Compare results from two independent approaches (e.g., `Counter` from the `collections` module  
vs. the manual `get()` loop) and assert they are equal.

**Pretty-print the output.**  
Use f-strings with alignment to make dictionary output readable:

```python
for word, count in sorted(histo_dict.items(), key=lambda x: -x[1])[:10]:
    print(f'{word:<20} {count:>4}')
```
